```python
#Decorators.
#A simple decorator to print before and after function
def log(func):
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper

@log
def greet():
    print("hello")

greet()
```

```python
#Decorator that accepts arguments.
def log(func):
    def wrapper(*args, **kwargs):
        print("Function called")
        return func(*args, **kwargs)
    return wrapper

@log
def add(a, b):
    return a + b

print(add(5, 3))
```

```python
#Decorator to measure execution time.

import time 

def timer(func):
    def wrapper(*args , **kwargs):
        start = time.time()
        result = func(*args , **kwargs)
        end = time.time()
        print("Time:", end - start)
        return result
    return wrapper

@timer
def calc():
    sum([i for i in range(5000)])

calc()
```

```python
#Decorator to restrict function if not authenticated.
def require_auth(func):
    def wrapper(user):
        if user != "admin":
            print("Access denied")
        else:
            func(user)
    return wrapper

@require_auth
def dashboard(user):
    print("Welcome", user)

dashboard("guest")
dashboard("admin")
```

```python
#Decorator chaining (two decorators).
def deco1(func):
    def wrapper():
        print("A")
        func()
    return wrapper

def deco2(func):
    def wrapper():
        print("B")
        func()
    return wrapper

@deco1
@deco2
def test():
    print("C")

test()
```

```python
# Generator.
# Create a generator for first n natural numbers.
def gen(n):
    for i in range(1,n+1):
        yield i

for x in gen(5):
    print(x)
```

```python
# fibonacci generator
def fibonacci(n):
    a,b = 0,1
    for _ in range(n):
        yield a 
        a,b = b , a+b

print(list(fibonacci(7)))
```

```python
# generator to yield squares of numbers
def squares(nums):
    for n in nums:
        yield n*n

print(list(squares([1, 2, 3, 4])))
```

```python
# infinite generator (natural numbers).
def infinite():
    n = 1
    while True:
        yield n
        n += 1

gen = infinite()
for _ in range(5):
    print(next(gen))
```

```python
# generator expression example
gen = (x*x for x in range(5))
print(list(gen))
```
