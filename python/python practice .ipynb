```python
# 8.remove duplicates 
arr=[1,2,3,3,4,4,5,6,7,7,8]
unique=[]
for x in arr:
    repeated=0
    for u in unique:
        if x == u:
            repeated=1
            break
            
    if repeated==0:
        unique.append(x)
        
print(unique)
```

```python
# 9.remove duplicates in dictionary (used set sunction)
d={"a":10, "b":20,"c":10}
unique={}
seen=set()
for k,v in d.items():
    if v not in seen:
        unique[k]=v
        seen.add(v)
print(unique)
```

```python
# 10.character frequency 
s="banana"
count={}
for char in s:
    if char not in count:
        count[char]=1
    else:
        count[char]+=1
print(count)
    
```

```python
#11. find missing no. from (1 to n) (no sum function)
arr=[1,2,4,5]
n=5
expected_sum=0
for i in range (1,n+1):
    expected_sum+=i
    actual_sum=0
    for x in arr:
        actual_sum+=x
print(expected_sum-actual_sum)
```

```python
# 12.maximun occuring character 
s="banana"
freq={}
for char in s:
    if char not in freq:
        freq[char]=1
    else:
        freq[char]+=1

max_char= None
max_count = 0 

for k in freq:
    if freq[k]>max_count:
        max_count=freq[k]
        max_char=k
print(max_char,max_count)
```

```python
#15.Intersection of two lists
a=[1,2,3,4]
b=[3,4,5,6]
inter=[]
for x in a:
    for y in b:
        if x==y:
            inter.append(x)
print(inter)
```

```python
# 16.pattern 
n=5
for i in range (1,n+1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
# 17 .pattern pyramid
n=5
for i in range (1,n+1):
    for j in range (n-i):
        print(" ",end=" ")    # printing space
    for j in range(2*i-1):
        print("*",end=" ")
    print()
```

```python
#18.check armstrong number i.e 153 = 1³ + 5³ + 3³
num=153
temp=num
total=0
while temp>0:
    digit=temp%10
    total=total+digit*digit*digit
    temp=temp//10
if total==num:
    print("armstrong")
else:
    print("not armstrong")
```

```python
#19 . check anagram .
s1="listen"
s2="silent"

if len(s1)!=len(s2):
    print(False)

else:
    is_anagram=True
    for ch in s1:
        count1=0
        count2=0
        for c in s1:
            if c==ch:
                count1 +=1
        for c in s2:
            if c==ch:
                count2 +=1

        if count1 != count2:
            is_anagram=False
    print(is_anagram)
```
