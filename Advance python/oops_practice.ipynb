```python
# public and private attributes
class Student:
    def __init__(self, name, marks):
        self.name = name         # public
        self._marks = marks      # protected
        self.__grade = "A"       # private

    def get_grade(self):
        return self.__grade

s = Student("Yash", 88)
print(s.name)
print(s._marks)
print(s.get_grade())

```

```python
# encapsulation
class Bank:
  def __init__(self, balance):
    self.__balance = balance

  def get_balance(self):
    return self.__balance

  def set_balance(self, amount):
    if amount > 0:
      self.__balance = amount
    else:
      print("Balance cannot be negative")

b = Bank(1000)
print(b.get_balance())
b.set_balance(5050)
print(b.get_balance())
```

```python
# encapsulation example 2
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self._balance = balance
        self.__pin = 1234

    def deposit(self, amount):
        self._balance += amount

    def show_balance(self):
        return self._balance

    def __show_pin(self):
        return self.__pin


acc = BankAccount("Yash", 1000)
acc.deposit(500)
print(acc.owner)
print(acc.show_balance())
# print(acc._BankAccount__show_pin())
```

```python
# abstraction
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        return "Car starts with a key"

class Bike(Vehicle):
    def start(self):
        return "Bike starts with self-start button"


print(Car().start())
print(Bike().start())
```

```python
# Inheritance - Single
class Parent :
  def show (self):
    print("this is parent class")
class child(Parent):
  pass

obj = child()
obj.show()
```

```python
# example 2 . single inheritance
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

d = Dog()
print(d.speak())
```

```python
# Multiple inheritance
class Father:
    def skills(self):
        return "Programming"

class Mother:
    def skills(self):
        return "Management"

class Child(Father, Mother):
    pass

c = Child()
print(c.skills())
```

```python
# Inheritance - Multilevel
class A:
  def showA(self):
    print("Aa")

class B(A):
  def showB(self):
      print("Bb")

class C(B):
  def showC(self):
      print("Cc")

obj = C()
obj.showA()
obj.showB()
obj.showC()
```

```python
# Polymorphism
class Animal:
    def sound(self):
        return "sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

animals = [Dog(), Cat()]
for a in animals:
    print(a.sound())
```

```python
# recurrsion
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

print(fact(5))
```

```python
# Question 1.
# Create a class Rectangle with methods to calculate area and perimeter.   constructor, instance attributes, method usage

class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
        return self.length*self.width

  def perimeter(self):
        return 2*(self.length+self.width)

r = Rectangle(4,5)
print(r.area())
print(r.perimeter())
```

```python
# Q4.Write code using encapsulation with getters and setters.  # private variable protection
class Student:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def set_name(self, name):
        if len(name)>1 :
          self.__name = name

s = Student("Yash")
print(s.get_name())
s.set_name("Aashu")
print(s.get_name())
```

```python
# Q5.Create a class Bank that allows deposit, withdraw, and check balance.
class Bank:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

b = Bank(1000)
b.deposit(500)
b.withdraw(200)
print(b.get_balance())

```
