# List Comprehension Practice 

```python
# 1. Squares from 1 to 10
[x**2 for x in range(1, 11)]
```

```python
# 2. Even numbers from 1 to 20
[x for x in range(1, 21) if x % 2 == 0]
```

```python
# 3. Uppercase strings
names = ["python", "java", "sql"]
[name.upper() for name in names]
```

```python
# 4. Length of each word
words = ["apple", "banana", "kiwi"]
[len(word) for word in words]
```

```python
# 5. Numbers greater than 10
nums = [5, 12, 7, 18]
[x for x in nums if x > 10]
```

```python
# 6. First letter of each word
[word[0] for word in ["data", "science", "ml"]]
```

```python
# 7. Convert strings to integers
nums = ["1", "2", "3"]
[int(x) for x in nums]
```

```python
# 8. Square only even numbers
[x**2 for x in range(1, 11) if x % 2 == 0]
```

```python
# 9. Replace negative numbers with 0
nums = [-1, 5, -3, 8]
[x if x > 0 else 0 for x in nums]
```

```python
# 10. Flatten 2D list
matrix = [[1,2],[3,4],[5,6]]
[item for sublist in matrix for item in sublist]
```

```python
# 11. Vowels from string
[x for x in "education" if x in "aeiou"]
```

```python
# 12. Number and square tuple
[(x, x**2) for x in range(1, 6)]
```

```python
# 13. Remove empty strings
lst = ["hi", "", "hello"]
[x for x in lst if x]
```

```python
# 14. Reverse strings
[word[::-1] for word in ["abc", "xyz"]]
```

```python
# 15. Divisible by 3
[x for x in range(1, 31) if x % 3 == 0]
```

```python
# 16. Boolean even check
[x % 2 == 0 for x in range(5)]
```

```python
# 17. Celsius to Fahrenheit
[c * 9/5 + 32 for c in [0, 20, 30]]
```

```python
# 18. Remove vowels
[x for x in "python" if x not in "aeiou"]
```

```python
# 19. Words longer than 4 chars
[word for word in ["cat", "tiger", "elephant"] if len(word) > 4]
```

```python
# 20. Number and cube
[(x, x**3) for x in range(1, 6)]
```
