```python
# reverse a string
string ='reverse'
rev=""
for char in string:
    rev = char + rev
print(rev)
```

```python
# check a string is palindrome or not
string = "yash"
rev=""
for char in string:
    rev = char + rev

if string == rev:
    print("palindrome")

else:
    print("not palindrome")

print(string)
print(rev)
```

```python
# palindrome using two pointer method
string = "radar"
left = 0 
right = len(string)-1

is_palindrome = True

while left < right :
    if string[left] != string[right]:
        is_palindrome = False
        break
    left +=1
    right-=1

if is_palindrome:
    print("palindrome")
else:
    print("not palindrome")
```

```python
# check is  two string is are Anagram
s1="silent"
s2="listen"

# check length , if not same , cannot be anagram
if len(s1) != len(s2):
    print("not anagram")
else:
    freq={}

    # count characters in first string 
    for ch in s1:
        if ch in freq:
            freq[ch] +=1
        else:
            freq[ch] = 1

    # decrease count using second string 
    is_anagram = True
    for ch in s2:
        if ch not in freq or freq[ch] == 0:
            is_anagram = False
            break
        freq[ch] -=1

    if is_anagram:
        print("Anagram")
    else:
        print("Not Anagram")
```
