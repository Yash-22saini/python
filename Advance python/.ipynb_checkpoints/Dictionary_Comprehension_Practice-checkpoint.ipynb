# Dictionary Comprehension Practice 

```python
# 1. Square values
nums = {1:2, 2:3}
{x: y**2 for x, y in nums.items()}
```

```python
# 2. Swap keys and values
d = {"a":1, "b":2}
{v: k for k, v in d.items()}
```

```python
# 3. Numbers and squares
{x: x**2 for x in range(1, 6)}
```

```python
# 4. Filter values > 10
d = {"a":5, "b":15}
{k: v for k, v in d.items() if v > 10}
```

```python
# 5. Uppercase values
d = {"name":"yash", "course":"ds"}
{k: v.upper() for k, v in d.items()}
```

```python
# 6. Word length mapping
words = ["apple", "banana"]
{word: len(word) for word in words}
```

```python
# 7. Even numbers only
{x: x**2 for x in range(10) if x % 2 == 0}
```

```python
# 8. Convert values to int
d = {"a":"1","b":"2"}
{k: int(v) for k, v in d.items()}
```

```python
# 9. From two lists
keys = ["a", "b"]
values = [1, 2]
{k: v for k, v in zip(keys, values)}
```

```python
# 10. Square even values
d = {"a":2, "b":3}
{k: v**2 for k, v in d.items() if v % 2 == 0}
```

```python
# 11. Character frequency
{s: "python".count(s) for s in "python"}
```

```python
# 12. Index mapping
{value: index for index, value in enumerate(["a","b","c"])}
```

```python
# 13. Cubes
{x: x**3 for x in range(1, 6)}
```

```python
# 14. Boolean mapping
{x: x > 5 for x in range(1, 10)}
```

```python
# 15. Remove empty values
d = {"a":"", "b":"hi"}
{k: v for k, v in d.items() if v}
```

```python
# 16. Word uppercase
{word: word.upper() for word in ["python", "sql"]}
```

```python
# 17. Remove negative values
d = {"a":-1, "b":3}
{k: v for k, v in d.items() if v > 0}
```

```python
# 18. ASCII values
{char: ord(char) for char in "abc"}
```

```python
# 19. Length of values
d = {"a":"hello", "b":"hi"}
{k: len(v) for k, v in d.items()}
```

```python
# 20. Even or odd mapping
{x: "even" if x % 2 == 0 else "odd" for x in range(5)}
```
