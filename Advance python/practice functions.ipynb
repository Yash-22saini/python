```python
# 1. Remove duplicates from a list without using set.
def remove_duplicates(lst):
    result = []
    for i in lst:
        if i not in result:
            result.append(i)
    return result
```

```python
# 2. Flatten a nested list
def flatten_list(lst):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(flatten_list(i))
        else:
            result.append(i)
    return result
```

```python
# 3. Rotate a list to the right by k steps
def rotate_list(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

# 4. Find all pairs with sum equal to target
def find_pairs(lst, target):
    result = []
    for i in range(len(lst)):
        for j in range(i+1, len(lst)):
            if lst[i] + lst[j] == target:
                result.append((lst[i], lst[j]))
    return result

# 5. Group values by key from list of tuples
def group_by_key(pairs):
    result = {}
    for key, value in pairs:
        if key in result:
            result[key].append(value)
        else:
            result[key] = [value]
    return result
```

```python
# 6. Return tuple with max sum
def max_sum_tuple(lst):
    return max(lst, key=lambda x: sum(x))

# 7. Word frequency in a sentence
def word_frequency(sentence):
    words = sentence.split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq

# 8. Merge two dictionaries and sum common keys
def merge_dicts(d1, d2):
    result = d1.copy()
    for key in d2:
        result[key] = result.get(key, 0) + d2[key]
    return result
```

```python
# 9. Sort dictionary by values in descending order
def sort_dict_desc(d):
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=True))

# 10. Invert a dictionary
def invert_dict(d):
    return {v: k for k, v in d.items()}
```

```python
# Case Study: Sales and Inventory Management System
class Product:
    def __init__(self, name, price, cost):
        self.name = name
        self.price = price
        self.cost = cost
        self.stock = 0

class Inventory:
    def __init__(self):
        self.products = {}

    def add_product(self, product, quantity):
        self.products[product.name] = product
        product.stock += quantity

    def update_stock(self, name, quantity):
        if name in self.products:
            self.products[name].stock += quantity

class SalesRecord:
    def __init__(self):
        self.sales = []

    def record_sale(self, product, quantity):
        self.sales.append((product.name, quantity, product.price, product.cost))

    def total_revenue(self):
        return sum(q * p for _, q, p, _ in self.sales)

    def total_profit(self):
        return sum(q * (p - c) for _, q, p, c in self.sales)

class StoreManager:
    def __init__(self):
        self.inventory = Inventory()
        self.sales_record = SalesRecord()

    def add_product(self, name, price, cost, quantity):
        product = Product(name, price, cost)
        self.inventory.add_product(product, quantity)

    def purchase(self, name, quantity):
        if name in self.inventory.products:
            product = self.inventory.products[name]
            if product.stock >= quantity:
                product.stock -= quantity
                self.sales_record.record_sale(product, quantity)

    def report(self):
        print("Revenue:", self.sales_record.total_revenue())
        print("Profit:", self.sales_record.total_profit())
```
