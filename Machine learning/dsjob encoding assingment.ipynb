```python
from sklearn.preprocessing import LabelEncoder , OrdinalEncoder , OneHotEncoder
```

```python
lb = LabelEncoder()
oe = OrdinalEncoder(categories = [['Male', 'Female']])
ohe = OneHotEncoder(drop = 'first', sparse_output = False)

df["gender"]=lb.fit_transform(df['gender'])
df["city_development_index"]=lb.fit_transform(df['city_development_index'])
df["enrolled_university"]=lb.fit_transform(df['enrolled_university'])
df["education_level"]=lb.fit_transform(df['education_level'])
```

```python
df['major_discipline']=lb.fit_transform(df['major_discipline'])
df['experience']=lb.fit_transform(df['experience'])
df['company_size']=lb.fit_transform(df['company_size'])
df["company_type"]=lb.fit_transform(df['company_type'])
df["last_new_job"]=lb.fit_transform(df['last_new_job'])
```
